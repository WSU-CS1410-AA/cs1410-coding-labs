## 1. Output

`print()` is the primary way to display output in Python. You can pass it any number
of arguments separated by commas — it will insert a space between each and add a
newline at the end. Expressions are evaluated before printing.

````python
print("Hello, World!")
print("Name:", "Alice", "Age:", 30)
print(1 + 2 + 3)              # 6 — expression evaluated first
print("Ten divided by three:", 10 / 3)
print(True, False, None)      # you can print any type
````

The default space separator and newline ending can both be changed using the
`sep` and `end` keyword arguments. This is useful for formatting tables or
building output that spans multiple `print()` calls:

````python
# sep controls what goes between arguments (default is a space)
print("2024", "01", "15", sep="-")      # 2024-01-15
print("Alice", "Bob", "Carol", sep=", ") # Alice, Bob, Carol

# end controls what comes after the last argument (default is newline)
print("loading", end="")
print(" . . . done")            # continues on the same line

# combining both
for i in range(1, 6):
    print(i, end=" ")           # 1 2 3 4 5  (all on one line)
print()                         # add newline at the end
````

**Escape sequences** embed special characters inside a string literal.
The most common ones:

| Sequence | Effect | Example output |
|---|---|---|
| `\n` | newline | moves to next line |
| `\t` | tab | aligns to next tab stop |
| `\\` | literal backslash | `\` |
| `\'` | single quote inside single-quoted string | `'` |
| `\"` | double quote inside double-quoted string | `"` |

````python
print("First line\nSecond line\nThird line")
print()
print("Name\t\tScore")
print("Alice\t\t95")
print("Bob\t\t88")
print()
print("She said, \"Python is fun!\"")
print('It\'s a beautiful day.')
print("Path: C:\\Users\\Alice\\Documents")
````

**f-strings** (formatted string literals, introduced in Python 3.6) are the
modern and most readable way to embed values in strings. Prefix the string
with `f` and place any expression inside `{}`:

````python
name  = "Alice"
age   = 20
score = 94.5

# Basic embedding
print(f"Hello, {name}! You are {age} years old.")

# Expressions inside braces are evaluated
print(f"In 10 years: {age + 10}")
print(f"Double score: {score * 2}")

# String methods work too
print(f"Uppercase: {name.upper()}")
````

**Format specifiers** after a colon inside `{}` control width, alignment, and
decimal precision. These concepts map directly to `setw()`, `setprecision()`,
`left`, and `right` in C++ `<iomanip>`:

````python
pi   = 3.14159265
n    = 42
name = "Alice"

print(f"{pi:.2f}")           # 3.14      — 2 decimal places
print(f"{pi:10.3f}")         # right-justified in width 10, 3 decimals
print(f"{n:08d}")            # 00000042  — zero-padded integer
print()

# Column alignment
print(f"{'Name':<12} {'Score':>8} {'Grade':^6}")
print("-" * 28)
print(f"{'Alice':<12} {95:>8} {'A':^6}")
print(f"{'Bob':<12} {82:>8} {'B':^6}")
print(f"{'Carol':<12} {73:>8} {'C':^6}")
````

## 2. User Input

`input()` pauses the program, displays an optional prompt string, reads one line
of text from the user, and **always returns it as a string** — regardless of what
the user types. You must convert explicitly when you need a number.

This is the most common beginner mistake: forgetting the conversion and then
wondering why arithmetic produces unexpected results.

````python
# Reading a string
name = input("Enter your name: ")
print(f"Hello, {name}!")
````

When the input needs to be used as a number, wrap `input()` in `int()` or `float()` before storing it:

````python
# Reading numbers — must convert explicitly
age    = int(input("How old are you? "))
height = float(input("Height in meters? "))

print(f"In 5 years you will be {age + 5} years old.")
print(f"Height in cm: {height * 100:.1f}")
print(f"Height in inches: {height * 39.37:.1f}")
````

Here is what happens when you forget to convert — the `+` operator on strings
does concatenation, not addition. Uncomment the last line to see the `TypeError`:

````python
x = input("Enter a number: ")
print(f"Type: {type(x)}")    # <class 'str'> — it is a string, not a number
print(x + x)                 # "55" if you typed 5 — string duplication, not 10
# print(x + 1)               # TypeError — uncomment to see the error
````

**A complete input/output example** — a temperature converter that reads from
the user and prints a formatted result:

````python
celsius = float(input("Enter temperature in Celsius: "))

fahrenheit = 9.0 * celsius / 5.0 + 32
kelvin     = celsius + 273.15

print()
print(f"  Celsius:    {celsius:8.2f} °C")
print(f"  Fahrenheit: {fahrenheit:8.2f} °F")
print(f"  Kelvin:     {kelvin:8.2f} K")
````

> **C++ note:** `cin >> x` reads directly into a typed variable, so the
> conversion is automatic based on the declared type. The equivalent of
> `age = int(input("How old are you? "))` in C++ is:
> ```cpp
> int age;
> cout << "How old are you? ";
> cin >> age;
> ```

## 3. Variables and Data Types

A variable is a named label that refers to a value stored in memory. In Python,
you create a variable by assigning a value — no type declaration is needed.
The **type belongs to the value**, not the variable name, which means you can
reassign a variable to a value of a completely different type (though this is
rarely a good idea in practice).

````python
count    = 10        # int   — whole numbers, positive or negative
price    = 3.99      # float — numbers with a decimal point
greeting = "Hello"   # str   — a sequence of characters
valid    = True      # bool  — either True or False
nothing  = None      # NoneType — represents "no value" / "not set"

# type() tells you the type of any value
print(type(count))
print(type(price))
print(type(greeting))
print(type(valid))
print(type(nothing))
````

**`None`** is a special value that means "nothing" or "not yet assigned". It is
Python's equivalent of `null` in some languages. You will encounter it as the
default return value of functions that do not explicitly return anything:

````python
def greet(name):
    print(f"Hello, {name}!")    # no return statement

result = greet("Alice")
print(result)                   # None — the function returned nothing

x = None
print(x is None)                # True — the idiomatic way to test for None
print(x is not None)            # False
````

**Multiple assignment** — assign several variables at once on a single line:

````python
x, y, z = 1, 2, 3
print(x, y, z)

# Swap two variables — no temp variable needed in Python
a, b = 10, 20
a, b = b, a
print(a, b)         # 20 10

# Assign the same value to several variables at once
p = q = r = 0
print(p, q, r)
````

**Type conversion** — Python does not convert types automatically in most cases.
Use these built-in functions:

| Function | Converts to | Behavior |
|---|---|---|
| `int(x)` | integer | truncates floats; parses numeric strings; `int(True)` → 1 |
| `float(x)` | float | widens integers; parses decimal strings |
| `str(x)` | string | works on any type |
| `bool(x)` | boolean | `False` for `0`, `0.0`, `""`, `None`, empty containers; `True` for everything else |

````python
# float → int truncates (does not round)
print(int(3.2))      # 3
print(int(3.9))      # 3   still 3, not 4
print(int(-3.7))     # -3  toward zero

# Use round() when you want rounding
print(round(3.9))    # 4
print(round(3.14159, 2))  # 3.14

# string → number
print(int("42"))
print(float("3.14"))

# anything → bool
print(bool(0), bool(1), bool(-5))    # False True True
print(bool(""), bool("hi"))          # False True
print(bool(None), bool([]))          # False False (empty list is falsy)
````

**Augmented assignment operators** update a variable in place:

````python
score = 100
score += 10     # score = score + 10  →  110
score -= 5      # score = score - 5   →  105
score *= 2      # score = score * 2   →  210
score //= 7     # score = score // 7  →  30
score %= 8      # score = score % 8   →  6
score **= 3     # score = score ** 3  →  216
print(score)

# Works on strings too
word = "hello"
word += " world"
print(word)

word *= 2
print(word)
````

## 4. Arithmetic and Math

Python supports the standard arithmetic operators. The ones that behave
differently than you might expect coming from everyday math are:
- `/` **always** produces a float, even when dividing two integers
- `//` performs floor division (rounds toward negative infinity)
- `%` gives the remainder after floor division
- `**` is exponentiation (there is no `^` operator for this in Python or C++)

````python
a, b = 17, 5

print(f"{a} +  {b} = {a +  b}")   # 22
print(f"{a} -  {b} = {a -  b}")   # 12
print(f"{a} *  {b} = {a *  b}")   # 85
print(f"{a} /  {b} = {a /  b}")   # 3.4   — always float
print(f"{a} // {b} = {a // b}")   # 3     — floor division
print(f"{a} %  {b} = {a %  b}")   # 2     — remainder
print(f"{a} ** {b} = {a ** b}")   # 1419857
````

The difference between `/` and `//` matters a lot:

````python
print(7 / 2)     # 3.5   — true division, always float
print(7 // 2)    # 3     — floor division, integer result
print(-7 // 2)   # -4    — floors toward negative infinity (not -3!)
print(7 % 2)     # 1     — remainder
print(-7 % 2)    # 1     — consistent with floor division

# Useful pattern: check even/odd with %
for n in range(8):
    label = "even" if n % 2 == 0 else "odd"
    print(f"{n} is {label}")
````

**Order of operations** follows standard math (PEMDAS/BODMAS).
Use parentheses to make intent explicit — it costs nothing and avoids bugs:

````python
print(2 + 3 * 4)        # 14 — * before +
print((2 + 3) * 4)      # 20 — parentheses first
print(10 / 2 + 1)       # 6.0
print(10 / (2 + 1))     # 3.333...
print(2 ** 3 ** 2)      # 512 — ** is right-to-left: 2 ** (3**2) = 2**9
print((2 ** 3) ** 2)    # 64
````

**Useful built-in math functions** — these are available without importing anything:

````python
print(abs(-42))          # 42  — absolute value
print(abs(3.7))          # 3.7
print(round(3.14159, 2)) # 3.14
print(round(2.5))        # 2   — Python rounds to even (banker's rounding)
print(round(3.5))        # 4
print(min(4, 7, 2, 9))   # 2
print(max(4, 7, 2, 9))   # 9
print(sum([1, 2, 3, 4, 5]))  # 15
print(pow(2, 10))        # 1024 — same as 2 ** 10
````

For more advanced math, import the `math` module. It provides constants and
functions you will also find in C++'s `<cmath>`:

````python
import math

print(math.pi)               # 3.14159...
print(math.e)                # 2.71828...
print(math.sqrt(144))        # 12.0
print(math.floor(3.9))       # 3   — round down
print(math.ceil(3.1))        # 4   — round up
print(math.log10(1000))      # 3.0
print(math.log(math.e))      # 1.0 — natural log
print(math.sin(math.pi / 2)) # 1.0
print(math.cos(0))           # 1.0

# Practical: area and circumference of a circle
r = 5.0
print(f"Area: {math.pi * r**2:.4f}")
print(f"Circumference: {2 * math.pi * r:.4f}")
````

**A practical example** — computing a grade average and letter grade:

````python
import math

# Three exam scores
e1, e2, e3 = 88, 73, 95
average = (e1 + e2 + e3) / 3

print(f"Scores:  {e1}, {e2}, {e3}")
print(f"Average: {average:.1f}")
print(f"Rounded: {round(average)}")

# Nearest 5-point boundary
boundary = math.floor(average / 5) * 5
print(f"Nearest 5-point floor: {boundary}")
````

> **C++ note:** C++ `/` on two integers does integer division (like Python `//`).
> To get a decimal result, at least one operand must be a float:
> `7 / 2` → `3` in C++, but `7.0 / 2` → `3.5`.
> C++ also has `<cmath>` with `sqrt()`, `floor()`, `ceil()`, `pow()`, `abs()`, `M_PI`.

## 5. Strings

A string is an **immutable** sequence of characters. "Immutable" means you cannot
change individual characters in place — instead, string operations always produce
a new string. You can use single quotes, double quotes, or triple quotes:

````python
s1 = 'Hello'
s2 = "World"
s3 = """This string
spans multiple lines
without any escape sequences."""

print(s1, s2)
print(s3)
print(len(s1))         # 5 — number of characters
print(type(s1))        # <class 'str'>
````

**Building strings** — concatenation with `+` and repetition with `*`:

````python
first = "Ada"
last  = "Lovelace"

full = first + " " + last
print(full)

print("-" * 30)                # a dividing line
print("ha" * 5)                # hahahahaha

# Building a string in a loop
result = ""
for i in range(1, 6):
    result += str(i) + " "     # append each number
print(result.strip())          # "1 2 3 4 5" 
````

### Indexing and slicing

Each character has a **positive index** starting at 0 and a **negative index**
starting at -1 from the end. A **slice** `s[start:stop:step]` extracts a range
of characters — `stop` is exclusive (the character at `stop` is not included):

````python
s = "Python"
#    0  1  2  3  4  5     positive indices
#   -6 -5 -4 -3 -2 -1     negative indices

print(s[0])       # P     first character
print(s[5])       # n     last character (by positive index)
print(s[-1])      # n     last character (by negative index)
print(s[-2])      # o     second-to-last

print(s[1:4])     # yth   indices 1, 2, 3 (stop=4 is excluded)
print(s[:3])      # Pyt   from start up to index 2
print(s[3:])      # hon   from index 3 to end
print(s[::2])     # Pto   every other character
print(s[::-1])    # nohtyP  reversed — classic Python trick
````

Since strings are immutable, you cannot change a character in place — trying to do so raises a `TypeError`. Instead, build a new string by concatenating the parts you want to keep:

````python
# Strings are immutable — you cannot assign to an index
word = "hello"
# word[0] = "H"   # TypeError — uncomment to see

# Instead, build a new string
word = "H" + word[1:]
print(word)  # Hello
````

### String methods

String methods always return a new string — the original is unchanged.
Here are the ones you will use most often:

````python
s = "  Hello, World!  "

print(s.strip())                             # "Hello, World!"     remove whitespace
print(s.strip().lstrip())                    # same — lstrip/rstrip remove one side only
print(s.strip().upper())                     # "HELLO, WORLD!"
print(s.strip().lower())                     # "hello, world!"
print(s.strip().title())                     # "Hello, World!"     capitalize each word
print(s.strip().replace("World", "Python"))  # "Hello, Python!"
print(s.strip().replace("l", "L"))          # replaces ALL occurrences
````

`split()` and `join()` are so often used as a pair that they are worth seeing together: `split` breaks a string into a list of parts, and `join` reassembles a list back into a string:

````python
# split and join — very commonly used together
sentence = "the quick brown fox"
words = sentence.split()             # split on whitespace by default
print(words)                         # ['the', 'quick', 'brown', 'fox']

csv = "Alice,88,A"
parts = csv.split(",")               # split on a specific delimiter
print(parts)                         # ['Alice', '88', 'A']

# join is the reverse of split
print(" ".join(words))               # "the quick brown fox"
print(", ".join(["a", "b", "c"]))   # "a, b, c" 
````

To locate or test for substrings, use `find()`, the `in` operator, `count()`, `startswith()`, and `endswith()` — no loop needed:

````python
# Searching
s = "Hello, World!"
print(s.find("World"))      # 7 — index of first match
print(s.find("Python"))     # -1 — not found
print("World" in s)         # True — membership test (preferred over find for True/False)
print(s.count("l"))         # 3 — how many times 'l' appears
print(s.startswith("Hello")) # True
print(s.endswith("!"))      # True
````

### Converting between strings and numbers

You often need to go back and forth between strings and numbers, especially
when reading from files or formatting output:

````python
# number → string
n = 42
s = str(n)
print(s, type(s))       # '42' <class 'str'>

# string → number
s = "3.14"
f = float(s)
print(f, type(f))       # 3.14 <class 'float'>

# building formatted output the long way (f-strings are usually better)
name  = "Alice"
score = 95
line  = name + ": " + str(score)
print(line)
````

**A practical string example** — reading and formatting a name:

````python
full_name = input("Enter your full name: ")

parts      = full_name.strip().split()
first      = parts[0].capitalize()
last       = parts[-1].upper()
initials   = ".".join(p[0].upper() for p in parts) + "."

print(f"First name:  {first}")
print(f"Last name:   {last}")
print(f"Initials:    {initials}")
print(f"Reversed:    {' '.join(reversed(parts))}")
````

## 6. Conditionals

Conditional statements let the program take different paths depending on whether
a condition is true or false. In Python the block belonging to each branch is
marked by **indentation** — everything at the same indent level is part of the
same block. The colon `:` at the end of each `if`, `elif`, and `else` line is
required.

````python
grade = 82

if grade >= 90:
    print("A — Excellent")
elif grade >= 80:
    print("B — Good")
elif grade >= 70:
    print("C — Satisfactory")
elif grade >= 65:
    print("D — Passing")
else:
    print("F — Failing")
````

**Comparison operators** evaluate to `True` or `False`:

| Operator | Meaning | Example |
|---|---|---|
| `==` | equal to | `x == 10` |
| `!=` | not equal to | `x != 0` |
| `<` / `>` | less / greater than | `x < 5` |
| `<=` / `>=` | less / greater than or equal | `x >= 18` |

Python also supports **chained comparisons**, which read like math notation:

````python
x = 7
print(x == 7)        # True
print(x != 5)        # True
print(x > 10)        # False
print(x <= 7)        # True

# Chained comparisons — unique to Python
print(1 < x < 10)    # True  — equivalent to (1 < x) and (x < 10)
print(0 <= x <= 5)   # False
````

**Logical operators** — `and`, `or`, `not` combine and negate conditions:

````python
age    = 20
has_id = True
is_vip = False

# and: both conditions must be True
if age >= 18 and has_id:
    print("Admitted")

# or: at least one condition must be True
if is_vip or age >= 21:
    print("VIP access")
else:
    print("Standard access")

# not: inverts a boolean
if not is_vip:
    print("Not a VIP")

# Short-circuit evaluation — Python stops as soon as the result is known
# 'and' stops at the first False; 'or' stops at the first True
x = 0
if x != 0 and 10 / x > 2:   # safe — second part never evaluated if x == 0
    print("big ratio")
else:
    print("x is zero or ratio is small")
````

**Nested conditionals** — place `if` statements inside other `if` blocks:

````python
x = float(input("Enter a number: "))

if x >= 0:
    if x == 0:
        print("Zero")
    elif x < 1:
        print("Between 0 and 1")
    else:
        print(f"Positive: {x}")
else:
    print(f"Negative: {x}")
````

**Conditional expression** (ternary operator) — a compact one-liner for
simple choices. The syntax is `value_if_true if condition else value_if_false`:

````python
x = 7
parity = "even" if x % 2 == 0 else "odd"
print(f"{x} is {parity}")

# Useful when assigning based on a condition
temperature = 35
weather = "hot" if temperature > 30 else ("cold" if temperature < 10 else "mild")
print(weather)
````

**FizzBuzz** — a classic conditional exercise:

````python
for n in range(1, 21):
    if n % 15 == 0:       # divisible by both 3 and 5 — check this first
        print("FizzBuzz")
    elif n % 3 == 0:
        print("Fizz")
    elif n % 5 == 0:
        print("Buzz")
    else:
        print(n)
````

> **C++ note:** C++ conditionals use the same logical operators (`&&` for `and`,
> `||` for `or`, `!` for `not`), but `if`/`else if`/`else` is the syntax (no `elif`).
> There is no chained comparison — `1 < x < 10` in C++ does not work as expected.

## 7. Loops

### `for` loop

A `for` loop iterates over any **sequence** — a range of numbers, a list,
a string, or other iterable. Use it when you know (or can compute) how many
iterations you need.

`range()` is the most common way to generate a sequence of integers:
- `range(n)` — 0 to n-1
- `range(start, stop)` — start to stop-1
- `range(start, stop, step)` — with a custom step (can be negative)

````python
# Basic counting
for i in range(5):
    print(i, end=" ")
print()

# From 1 to 10
for i in range(1, 11):
    print(i, end=" ")
print()

# Even numbers
for i in range(0, 20, 2):
    print(i, end=" ")
print()

# Countdown
for i in range(10, 0, -1):
    print(i, end=" ")
print("Blast off!")
````

Looping directly over a list or string:

````python
planets = ["Mercury", "Venus", "Earth", "Mars", "Jupiter"]
for planet in planets:
    print(planet)

print()

# Loop over a string — iterates character by character
for ch in "Python":
    print(ch, end=" ")
print()
````

`enumerate()` provides both the index and the value on each pass — use it
whenever you need the position alongside the element:

````python
fruits = ["apple", "banana", "cherry", "date"]

# Without enumerate — manually tracking the index
for i in range(len(fruits)):
    print(f"{i}: {fruits[i]}")
````

With `enumerate()` this becomes cleaner — you get the index and value paired together without manually indexing the list:

````python
fruits = ["apple", "banana", "cherry", "date"]

for i, fruit in enumerate(fruits):
    print(f"{i}: {fruit}")
````

Pass `start=` to begin the counter at any value. Starting at 1 is common for human-readable output:

````python
fruits = ["apple", "banana", "cherry", "date"]

for i, fruit in enumerate(fruits, start=1):
    print(f"{i}. {fruit}")
````

`zip()` lets you iterate over two sequences in parallel:

````python
names  = ["Alice", "Bob", "Carol"]
scores = [92, 85, 78]

for name, score in zip(names, scores):
    grade = "pass" if score >= 65 else "fail"
    print(f"{name:<10} {score:3}  {grade}")
````

**Nested loops** — the inner loop completes fully for each iteration of the outer:

````python
# Multiplication table
print("   ", end="")
for j in range(1, 6):
    print(f"{j:4}", end="")
print()
print("    " + "-" * 20)

for i in range(1, 6):
    print(f"{i:2} |", end="")
    for j in range(1, 6):
        print(f"{i*j:4}", end="")
    print()
````

**Computing statistics with a `for` loop:**

````python
numbers = [14, 27, 8, 35, 19, 42, 6, 31]

total   = 0
minimum = numbers[0]
maximum = numbers[0]

for n in numbers:
    total   += n
    if n < minimum:
        minimum = n
    if n > maximum:
        maximum = n

count   = len(numbers)
average = total / count

print(f"Count:   {count}")
print(f"Sum:     {total}")
print(f"Min:     {minimum}")
print(f"Max:     {maximum}")
print(f"Average: {average:.2f}")
````

### `while` loop

Use `while` when you do not know the number of iterations in advance — keep
going as long as a condition holds. **Always make sure something inside the
loop eventually makes the condition False** — otherwise you have an infinite loop.

````python
# Counting up
i = 1
while i <= 5:
    print(i, end=" ")
    i += 1
print()

# Fibonacci sequence up to 1000
a, b = 0, 1
while b <= 1000:
    print(b, end=" ")
    a, b = b, a + b
print()
````

The `while True` / `break` pattern is the standard way to write a **validation loop** — keep prompting until the input passes your check:

````python
# Keep asking until the user enters a valid positive number
while True:
    s = input("Enter a positive number: ")
    if s.lstrip("-").isdigit() and int(s) > 0:
        n = int(s)
        break
    print("  Invalid — please try again.")

print(f"You entered: {n}")
print(f"Its square:  {n ** 2}")
````

### `break` and `continue`

## 8. Lists

A **list** is an ordered, **mutable** sequence of values enclosed in `[]`.
It is the most-used data structure in Python. You can store any mix of types,
but in practice you will usually keep all elements the same type — just like
an array or `std::vector` in C++.

````python
# Creating lists
numbers = [10, 20, 30, 40, 50]
names   = ["Alice", "Bob", "Carol"]
empty   = []
mixed   = [1, "hello", 3.14, True]  # possible but unusual

print(numbers)
print(len(numbers))     # 5 — number of elements
print(numbers[0])       # 10 — first element
print(numbers[-1])      # 50 — last element
print(numbers[1:4])     # [20, 30, 40] — slice (exclusive end)
print(numbers[::-1])    # [50, 40, 30, 20, 10] — reversed
````

**Modifying lists** — unlike strings, lists are mutable:

````python
scores = [85, 92, 78, 95, 88]
print("Original:", scores)

# Direct index assignment changes an element in place
scores[2] = 80
print("After scores[2]=80:", scores)
````

`append()`, `insert()`, `remove()`, and `pop()` all modify the list in place. Each line below builds on the previous state:

````python
scores = [85, 92, 78, 95, 88]

scores.append(91)           # add to the end
print("After append(91):", scores)

scores.insert(1, 99)        # insert 99 at index 1
print("After insert(1,99):", scores)

scores.remove(99)           # remove first occurrence of value 99
print("After remove(99):", scores)

popped = scores.pop()       # remove and return the last element
print(f"Popped: {popped}")
print("After pop():", scores)

popped_at = scores.pop(0)   # remove and return element at index 0
print(f"Popped at 0: {popped_at}")
print("After pop(0):", scores)
````

**Searching and sorting:**

````python
nums = [4, 1, 7, 2, 9, 3, 8, 5, 6]

print("Contains 7:", 7 in nums)
print("Contains 10:", 10 in nums)
print("Count of 7:", nums.count(7))
print("Index of 9:", nums.index(9))     # position of first occurrence

print("Sum:", sum(nums))
print("Min:", min(nums))
print("Max:", max(nums))
````

`.sort()` reorders the list in place. When you need the original preserved, use the built-in `sorted()` function — it returns a new list and leaves the original unchanged:

````python
nums = [4, 1, 7, 2, 9, 3, 8, 5, 6]

nums.sort()                              # sort in place — modifies the list
print("Sorted:", nums)

nums.sort(reverse=True)                  # descending
print("Descending:", nums)

original = [4, 1, 7, 2]
sorted_copy = sorted(original)           # returns a new list — original unchanged
print("Original:", original)
print("Sorted copy:", sorted_copy)
````

**List comprehensions** — a concise way to build a list from another sequence:

````python
# squares of 1–10
squares = [x ** 2 for x in range(1, 11)]
print(squares)

# even numbers 0–19 — add an if-clause to filter
evens = [x for x in range(20) if x % 2 == 0]
print(evens)
````

The same syntax works with any iterable, not just `range`. You can transform strings, filter by a condition, or combine both:

````python
# transform each element
words = ["hello", "world", "python"]
upper = [w.upper() for w in words]
print(upper)

# filter: keep only passing scores, then report the class average
scores  = [85, 42, 91, 67, 38, 73, 95, 55]
passing = [s for s in scores if s >= 65]
print(f"Passing scores: {passing}")
print(f"Class average: {sum(scores)/len(scores):.1f}")
````

**2D lists** — a list of lists, useful for grids and tables:

````python
# A 3x3 grid
grid = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

# Access with two indices: grid[row][col]
print(grid[0][0])   # 1 — top-left
print(grid[1][2])   # 6 — row 1, col 2
print(grid[2][-1])  # 9 — bottom-right

# Iterate over a 2D list
for row in grid:
    for val in row:
        print(f"{val:3}", end="")
    print()
````

## 9. Dictionaries

A **dictionary** stores key-value pairs — like a real dictionary where you look
up a definition by its word. Keys must be unique and immutable (strings and
numbers work well as keys). Dictionaries are unordered in older Python but
maintain insertion order since Python 3.7.

In C++, the equivalent is `std::map<K, V>` or `std::unordered_map<K, V>`.

````python
# Creating a dictionary with curly braces
student = {
    "name":   "Alice",
    "age":    20,
    "grade":  "B",
    "score":  85.5
}

# Accessing values by key
print(student["name"])
print(student["score"])

# Adding and updating entries
student["year"] = 2
student["score"] = 88.0     # update existing key

print(student)
print(len(student))         # number of key-value pairs
````

Accessing a key that does not exist with `d[key]` raises a `KeyError`. Use `.get()` when the key might be absent — it returns `None` by default, or a custom default you supply:

````python
# Safe access with .get() — returns None (or a default) if key is missing
d = {"x": 10, "y": 20}

print(d.get("x"))           # 10
print(d.get("z"))           # None — no KeyError
print(d.get("z", 0))        # 0 — custom default

# Check membership
print("x" in d)             # True
print("z" in d)             # False
````

Dictionaries support three iteration views: the default (keys only), `.values()`, and `.items()` which yields both key and value as a tuple — the most useful form:

````python
# Iterating over a dictionary
scores = {"Alice": 92, "Bob": 85, "Carol": 78, "Dave": 95}

# Keys only
for name in scores:
    print(name)

print()

# Values only
for score in scores.values():
    print(score)

print()

# Both key and value — the most common pattern
for name, score in scores.items():
    grade = "A" if score >= 90 else "B" if score >= 80 else "C"
    print(f"{name:<10} {score:3}  {grade}")
````

A common pattern is building a dictionary dynamically — for example, counting how many times each item appears in a list:

````python
# Building a dictionary from a list — counting occurrences
words = ["apple", "banana", "apple", "cherry", "banana", "apple"]
counts = {}

for word in words:
    if word in counts:
        counts[word] += 1
    else:
        counts[word] = 1

print(counts)

# Sorted by count (descending)
for word, count in sorted(counts.items(), key=lambda kv: kv[1], reverse=True):
    print(f"{word:<10} {count}")
````

## 10. Functions

A **function** is a named, reusable block of code. Defining a function does
not run it — it just gives the code a name. You run it by **calling** it with
parentheses.

Good functions:
- Do **one** thing clearly
- Receive inputs as **parameters**
- Communicate results through **`return`** rather than by printing or modifying globals
- Have a descriptive name that says what they do

````python
# Defining and calling a function
def greet(name):
    print(f"Hello, {name}! Welcome to CS 1410.")

greet("Alice")
greet("Bob")
greet("Carol")
````

`return` sends a value back to the caller and immediately exits the function:

````python
def square(x):
    return x * x

def circle_area(radius):
    import math
    return math.pi * radius ** 2

print(square(5))                # 25
print(square(12))               # 144
print(f"{circle_area(3):.4f}")  # 28.2743

# The return value can be used in any expression
total = square(3) + square(4)
print(f"3² + 4² = {total}")     # 25  (a Pythagorean triple check)
````

**Multiple return values** — Python returns them as a tuple:

````python
def min_max_avg(numbers):
    """Return the minimum, maximum, and average of a list."""
    return min(numbers), max(numbers), sum(numbers) / len(numbers)

data = [14, 27, 8, 35, 19, 42, 6, 31]
lo, hi, avg = min_max_avg(data)
print(f"Min: {lo}, Max: {hi}, Average: {avg:.2f}")
````

**Default parameter values** make some arguments optional:

````python
def power(base, exponent=2):
    """Raise base to exponent (default: squared)."""
    return base ** exponent

def repeat(text, times=3, sep=" "):
    """Repeat text a given number of times with a separator."""
    return sep.join([text] * times)

print(power(4))          # 16   — default exponent 2
print(power(4, 3))       # 64
print(power(2, 10))      # 1024

print(repeat("ha"))              # ha ha ha
print(repeat("ho", 4))           # ho ho ho ho
print(repeat("*", 5, ""))        # *****
````

**Docstrings** document what a function does, what its parameters are, and
what it returns. They show up in `help()` and IDE tooltips — a good habit to
build before you reach C++, where header file comments serve the same purpose:

````python
def celsius_to_fahrenheit(celsius):
    """
    Convert a temperature from Celsius to Fahrenheit.

    Parameters:
        celsius (float): temperature in degrees Celsius

    Returns:
        float: temperature in degrees Fahrenheit
    """
    return 9.0 * celsius / 5.0 + 32

help(celsius_to_fahrenheit)
print(celsius_to_fahrenheit(0))    # 32.0
print(celsius_to_fahrenheit(100))  # 212.0
print(celsius_to_fahrenheit(-40))  # -40.0 — the one temperature equal in both scales
````

**Functions calling other functions** — build complex behavior from simple pieces:

````python
def is_leap_year(year):
    """Return True if year is a leap year."""
    return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)

print(is_leap_year(2024))   # True  — divisible by 4, not by 100
print(is_leap_year(1900))   # False — divisible by 100 but not 400
print(is_leap_year(2000))   # True  — divisible by 400
````

`days_in_month` calls `is_leap_year` to handle February correctly — functions can call other functions freely:

````python
def is_leap_year(year):
    return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)

def days_in_month(month, year):
    """Return the number of days in a given month and year."""
    thirty_one = [1, 3, 5, 7, 8, 10, 12]
    if month in thirty_one:
        return 31
    elif month == 2:
        return 29 if is_leap_year(year) else 28
    else:
        return 30

print(days_in_month(2, 2024))   # 29 — leap year February
print(days_in_month(2, 2023))   # 28 — non-leap year February
print(days_in_month(1, 2024))   # 31 — January
````

Summing `days_in_month` across all 12 months gives the total days in any year — no hardcoding 365 or 366:

````python
def is_leap_year(year):
    return (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0)

def days_in_month(month, year):
    thirty_one = [1, 3, 5, 7, 8, 10, 12]
    if month in thirty_one: return 31
    elif month == 2: return 29 if is_leap_year(year) else 28
    else: return 30

def days_in_year(year):
    """Return the total number of days in a year."""
    return sum(days_in_month(m, year) for m in range(1, 13))

print(days_in_year(2024))   # 366
print(days_in_year(2023))   # 365
````

### Scope — local vs. global variables

A variable created **inside** a function is **local** to that function — it
does not exist outside. A variable created **outside** all functions is
**global** and readable everywhere, but to **modify** it inside a function
you must declare it `global`. In practice, prefer passing values as parameters
and returning results rather than relying on global variables.

````python
x = "I am global"

def show_local():
    x = "I am local"          # creates a NEW local variable; global x is untouched
    print("inside:", x)

show_local()
print("outside:", x)          # unchanged
````

If a function needs to *modify* a global variable, declare it `global` inside the function. This is generally a sign of poor design — we will see the better alternative right after:

````python
total = 0                      # global accumulator

def add(n):
    global total
    total += n

add(10)
add(25)
add(7)
print(total)                   # 42
````

The better design is to avoid globals entirely — pass state as arguments:

````python
def add_to(total, n):
    return total + n           # pure function — no side effects

t = 0
t = add_to(t, 10)
t = add_to(t, 25)
t = add_to(t, 7)
print(t)                       # 42 — same result, no global needed
````

## 11. Putting It All Together

This final section combines everything — variables, strings, lists, dictionaries,
conditionals, loops, and functions — in one small but complete program.

The program reads a list of student names and scores, computes statistics,
assigns letter grades, and prints a formatted report:

````python
import math

def letter_grade(score):
    """Return the letter grade for a numeric score (0–100)."""
    if   score >= 90: return "A"
    elif score >= 80: return "B"
    elif score >= 70: return "C"
    elif score >= 65: return "D"
    else:             return "F"

def grade_point(letter):
    """Convert a letter grade to a GPA point value."""
    return {"A": 4.0, "B": 3.0, "C": 2.0, "D": 1.0, "F": 0.0}.get(letter, 0.0)

def statistics(values):
    """Return (min, max, mean, std_dev) for a list of numbers."""
    n    = len(values)
    mean = sum(values) / n
    variance = sum((v - mean) ** 2 for v in values) / n
    return min(values), max(values), mean, math.sqrt(variance)
````

The data is a dictionary mapping each student's name to a list of their exam scores:

````python
roster = {
    "Alice":   [88, 92, 79, 95],
    "Bob":     [72, 68, 75, 70],
    "Carol":   [95, 98, 91, 97],
    "Dave":    [60, 55, 63, 58],
    "Eve":     [84, 87, 80, 90],
}
````

Loop over the roster, compute each student's average, grade, and GPA, then sort by average so the top performer appears first:

````python
results = []
for name, scores in roster.items():
    avg   = sum(scores) / len(scores)
    grade = letter_grade(avg)
    gpa   = grade_point(grade)
    results.append({"name": name, "avg": avg, "grade": grade, "gpa": gpa})

results.sort(key=lambda r: r["avg"], reverse=True)   # sort by average descending
````

Finally, print the formatted report and compute class-wide statistics using the `statistics` helper:

````python
header = f"{'Name':<10} {'Average':>8} {'Grade':>6} {'GPA':>5}"
print(header)
print("-" * len(header))

for r in results:
    print(f"{r['name']:<10} {r['avg']:>8.1f} {r['grade']:>6} {r['gpa']:>5.1f}")

all_avgs = [r["avg"] for r in results]
lo, hi, mean, sd = statistics(all_avgs)

print("-" * len(header))
print(f"\nClass statistics ({len(results)} students):")
print(f"  Highest average : {hi:.1f}")
print(f"  Lowest average  : {lo:.1f}")
print(f"  Class mean      : {mean:.1f}")
print(f"  Std deviation   : {sd:.1f}")

passing = [r for r in results if r["grade"] != "F"]
print(f"  Passing         : {len(passing)} / {len(results)}")
````
